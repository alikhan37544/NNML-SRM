```python
from sklearn.mixture import GaussianMixture
from sklearn.metrics import silhouette_score


data = pd.read_csv('EM data.csv')

# Perform one-hot encoding for categorical features
data_encoded = pd.get_dummies(data, columns=['Location', 'Christmas', 'Newyear', 'Ramzan', 'Diwali'])

# Extract features for clustering
X = data_encoded.drop('CustomerID', axis=1) 

# Extract features for clustering (Location and Celebration behavior)
# X = df[['Location', 'Christmas', 'Newyear', 'Ramzan', 'Diwali']]

# Initialize GMM model
gmm = GaussianMixture(n_components=3, random_state=42)  # Assuming 3 clusters

# Fit the model to the data
gmm.fit(X)

# Predict cluster labels
cluster_labels = gmm.predict(X)

# Evaluate clustering quality
silhouette_avg = silhouette_score(X, cluster_labels)

# Print silhouette score
print("Silhouette Score:", silhouette_avg)

# Get parameters of the model
cluster_means = gmm.means_
cluster_covariances = gmm.covariances_
cluster_weights = gmm.weights_

# Print mean, variance, and cluster weights for each cluster
for i in range(len(cluster_means)):
    print("Cluster", i+1)
    print("Mean:", cluster_means[i])
    print("Covariance:", cluster_covariances[i])
    print("Weight:", cluster_weights[i])
    print()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

# Apply PCA to reduce the data to 2 dimensions
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the clustered data
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=cluster_labels, cmap='viridis', s=50, alpha=0.5)
plt.title('Gaussian Mixture Model Clustering')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar(label='Cluster Label')
plt.grid(True)
plt.show()
```

```python
import numpy as np

def E_step(X, cluster_means, cluster_covariances, cluster_weights):
    """
    Perform the E-step of the EM algorithm for GMMs.
    
    Parameters:
    - X: Input data matrix (shape: [n_samples, n_features]).
    - cluster_means: List of cluster means (shape: [n_clusters, n_features]).
    - cluster_covariances: List of cluster covariance matrices (shape: [n_clusters, n_features, n_features]).
    - cluster_weights: List of cluster weights.
    
    Returns:
    - gamma: Posterior probabilities (shape: [n_samples, n_clusters]).
    """
    n_samples = X.shape[0]
    n_clusters = len(cluster_means)
    gamma = np.zeros((n_samples, n_clusters))
    
    for k in range(n_clusters):
        mean = cluster_means[k]
        covariance = cluster_covariances[k]
        weight = cluster_weights[k]
        
        # Compute the probability density function (PDF) for each sample in cluster k
        # using the multivariate normal distribution
        pdf = (1 / np.sqrt((2 * np.pi) ** X.shape[1] * np.linalg.det(covariance))) * \
              np.exp(-0.5 * np.sum((X - mean) @ np.linalg.inv(covariance) * (X - mean), axis=1))
        
        # Compute the posterior probability (gamma) for each sample in cluster k
        gamma[:, k] = weight * pdf
    
    # Normalize the posterior probabilities
    gamma /= np.sum(gamma, axis=1, keepdims=True)
    
    return gamma

def M_step(X, gamma):
    """
    Perform the M-step of the EM algorithm for GMMs.
    
    Parameters:
    - X: Input data matrix (shape: [n_samples, n_features]).
    - gamma: Posterior probabilities (shape: [n_samples, n_clusters]).
    
    Returns:
    - cluster_means: Updated list of cluster means (shape: [n_clusters, n_features]).
    - cluster_covariances: Updated list of cluster covariance matrices (shape: [n_clusters, n_features, n_features]).
    - cluster_weights: Updated list of cluster weights.
    """
    n_samples = X.shape[0]
    n_features = X.shape[1]
    n_clusters = gamma.shape[1]
    
    # Compute the total posterior probability assigned to each cluster
    N_k = np.sum(gamma, axis=0)
    
    # Update cluster weights
    cluster_weights = N_k / n_samples
    
    # Update cluster means
    cluster_means = np.dot(gamma.T, X) / N_k[:, np.newaxis]
    
    # Update cluster covariances
    cluster_covariances = np.zeros((n_clusters, n_features, n_features))
    for k in range(n_clusters):
        # Compute the weighted difference between each sample and the cluster mean
        diff = X - cluster_means[k]
        
        # Compute the weighted outer product of the differences
        weighted_diff_outer = np.einsum('ij,ik->ijk', diff, diff)
        
        # Compute the weighted sum of the outer products
        weighted_sum = np.sum(gamma[:, k, np.newaxis, np.newaxis] * weighted_diff_outer, axis=0)
        
        # Divide by the total posterior probability for the cluster
        cluster_covariances[k] = weighted_sum / N_k[k]
    
    return cluster_means, cluster_covariances, cluster_weights

# Example usage:
# Assuming you have fitted a GMM model and obtained initial parameters (cluster_means, cluster_covariances, cluster_weights)
# Perform one iteration of the EM algorithm
gamma = E_step(X, cluster_means, cluster_covariances, cluster_weights)
cluster_means_new, cluster_covariances_new, cluster_weights_new = M_step(X, gamma)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.mixture import GaussianMixture

# Generate synthetic data
X, _ = make_blobs(n_samples=300, centers=3, random_state=42)

# Initialize GMM model
gmm = GaussianMixture(n_components=3, random_state=42)

# Perform one iteration of the EM algorithm
gamma = gmm.fit(X).predict_proba(X)
cluster_means_new, cluster_covariances_new, cluster_weights_new = M_step(X, gamma)

# Plot original data
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.scatter(X[:, 0], X[:, 1], c=np.argmax(gamma, axis=1), cmap='viridis', marker='o')
plt.title('Original Data with Cluster Assignments')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')

# Plot updated cluster means
plt.subplot(1, 2, 2)
plt.scatter(X[:, 0], X[:, 1], c=np.argmax(gamma, axis=1), cmap='viridis', marker='o')
plt.scatter(cluster_means_new[:, 0], cluster_means_new[:, 1], c='red', marker='x', s=100, label='New Means')
plt.title('Original Data with Updated Cluster Means')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

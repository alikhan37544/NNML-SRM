```python
import pandas as pd

# Load the dataset
data = pd.read_csv("EM data.csv")

# Display the first few rows of the dataset
print(data.head())
```

Calculate total amount spent by each customer over the given period

```python
total_amount_by_customer = data.groupby('CustomerID')['TotalAmount'].sum()
print(total_amount_by_customer)
```

Determine Frequency of Transactions for each customer

```python
transaction_frequency = data.groupby('CustomerID')['NumTransactions'].sum()
print(transaction_frequency)
```

Calculate average amount spent per transaction by each customer

```python
average_amount_per_transaction = total_amount_by_customer / transaction_frequency
print(average_amount_per_transaction)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('EM data.csv')

# Convert timestamp column to datetime format
data['Timestamp'] = pd.to_datetime(data['Timestamp'])

# Extract month and day from the timestamp
data['Month'] = data['Timestamp'].dt.month
data['Day'] = data['Timestamp'].dt.day

# Plot the distribution of transactions over time
plt.figure(figsize=(10, 6))
plt.plot(data['Month'], data['NumTransactions'], marker='o', linestyle='-')
plt.xlabel('Month')
plt.ylabel('Number of Transactions')
plt.title('Distribution of Transactions Over Time')
plt.xticks(range(1, 13))  # Set x-axis ticks to represent months
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame from the provided data
data = {
    'CustomerID': [1, 2, 3, 4, 5],
    'TotalAmount': [450, 120, 780, 320, 980],
    'NumTransactions': [5, 2, 6, 3, 7],
    'UniqueItems': [4, 2, 5, 3, 6],
    'Location': ['Texas', 'Texas', 'Texas', 'Texas', 'Texas'],
    'Christmas': ['yes', 'No', 'yes', 'yes', 'yes'],
    'Newyear': ['yes', 'No', 'yes', 'yes', 'yes'],
    'Ramzan': ['yes', 'No', 'No', 'No', 'No'],
    'Diwali': ['yes', 'No', 'No', 'No', 'No']
}

df = pd.DataFrame(data)

# Plot histogram for Total Amount spent
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.hist(df['TotalAmount'], bins=10, color='skyblue', edgecolor='black')
plt.title('Total Amount Spent Distribution')
plt.xlabel('Total Amount Spent')
plt.ylabel('Frequency')

# Plot histogram for Number of Transactions
plt.subplot(1, 2, 2)
plt.hist(df['NumTransactions'], bins=5, color='salmon', edgecolor='black')
plt.title('Number of Transactions Distribution')
plt.xlabel('Number of Transactions')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()

# Plot bar plots for customer segmentation based on Location and Celebration behavior
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
df['Location'].value_counts().plot(kind='bar', color='lightgreen')
plt.title('Customer Segmentation by Location')
plt.xlabel('Location')
plt.ylabel('Number of Customers')

plt.subplot(1, 2, 2)
celebrations = ['Christmas', 'Newyear', 'Ramzan', 'Diwali']
celebration_counts = df[celebrations].apply(pd.value_counts).T
celebration_counts.plot(kind='bar', stacked=True, color=['skyblue', 'salmon', 'lightgreen', 'orange'])
plt.title('Customer Segmentation by Celebration Behavior')
plt.xlabel('Celebration')
plt.ylabel('Number of Customers')

plt.tight_layout()
plt.show()
```
